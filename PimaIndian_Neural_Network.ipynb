### Introduction

In this Notebook we will pick up a dataset of interest to expolore data analysis and Machine Learning.

We have used the Pima Indian Diabetes dataset from the National Institute of Diabetes and Digestive and Kidney Diseases. The link to the dataset can be found in the reference section.

The problem at hand is to be able to detect and predict whether a person has Diabetes or not, based on other available parameters like Body Mass Index, Insulin levels, etc. This is a classification problem and would require a machine learning model to predict Diabetes.

Another interesting insight from this problem could be to see on which parameter does Diabetes depend the most.


#### Loading the data from local drive:

```
#import necessary models
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
import os
```

```
#No data cleaning done in this code.
dataset = pd.read_csv('/content/diabetes.csv')
X = dataset.iloc[:,0:8].values
y = dataset.iloc[:,8].values
#print (X)
```

```
dataset
```

```
# Display statistical summary of all numeric features
# (count, mean, std, min, quartiles, max)
dataset.describe()
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```
#Define Keras model
model = Sequential()
#Define input layer, first hidden layer and activation function
model.add(Dense(12, input_dim=8, activation='relu'))
#Define second hidden layer and activation function
model.add(Dense(8, activation= 'relu'))
#Define output layer and activation function, sigmoid activation since two classes
model.add(Dense(1, activation='sigmoid'))
```

```
model.summary()
```

```
#Compile Keras model
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics =['accuracy'])
```

```
#Fit keras model
model.fit(X,y, epochs=200, batch_size=10, verbose = 1)
```

```
#Evaluate the accuracy
_, accuracy = model.evaluate(X_train,y_train, verbose= 0)
print ('Accuracy: %.2f' % (accuracy*100) + '%')
```

```
# Evaluate the model on the test data using `evaluate`
print("Evaluate on test data")
results = model.evaluate(X_test, y_test, batch_size=128)
print("test loss, test acc:", results)
```

```
#Save the model
model.save('PimaIndian_model.h5')
```

```
#Inference Code
import numpy as np
from keras.models import load_model

# returns a compiled model
# identical to the previous one
model = load_model('PimaIndian_model.h5')
data= np.array([[6.   , 148.   ,  72.   ,  35.   ,   0.   ,  33.6  ,   0.627,
        50. ]])
result=(model.predict(data) > 0.5)
print(result[0][0])
```
